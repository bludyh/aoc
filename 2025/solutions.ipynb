### #1

```python
%%time

input_file = "1.txt"

dial_pointer = 50

zero_count = 0

def move_dial(dial_pointer, move):
    dial_pointer = (dial_pointer + move) % 100
    return dial_pointer

with open(input_file, "r") as file:
    for line in file:
        if line.startswith("R"):
            move = int(line[1:])
        elif line.startswith("L"):
            move = -int(line[1:])
        dial_pointer = move_dial(dial_pointer, move)

        if dial_pointer == 0:
            zero_count += 1

zero_count
```

### #2

```python
%%time

input_file = "1.txt"
dial_pointer = 50
total_zero_count = 0

def move_dial_count_zero(dial_pointer, move):
    move_value = dial_pointer + move

    if move > 0:
        if dial_pointer == 0:
            zero_count = len([i for i in range(dial_pointer, move_value + 1) if i % 100 == 0]) - 1
        else:
            zero_count = len([i for i in range(dial_pointer, move_value + 1) if i % 100 == 0])
    else:
        if dial_pointer == 0:
            zero_count = len([i for i in range(dial_pointer, move_value - 1, -1) if i % 100 == 0]) - 1
        else:
            zero_count = len([i for i in range(dial_pointer, move_value - 1, -1) if i % 100 == 0])

    return move_value % 100, zero_count

with open(input_file, "r") as file:
    for line in file:
        if line.startswith("R"):
            move = int(line[1:])
        elif line.startswith("L"):
            move = -int(line[1:])
        dial_pointer, zero_count = move_dial_count_zero(dial_pointer, move)
        total_zero_count += zero_count

total_zero_count
```

```python
%%time

input_file = "2.txt"

invalid_sum = 0

def has_repeated_sequence(number):
    num_str = str(number)
    length = len(num_str)

    if length % 2 != 0:
        return False

    first_half = num_str[: length // 2]
    second_half = num_str[length // 2 :]

    return first_half == second_half

with open(input_file, "r") as file:
    ranges = file.read().strip().split(",")

    for id_range in ranges:
        start = int(id_range.split("-")[0])
        end = int(id_range.split("-")[1])

        for i in range(start, end + 1):
            if has_repeated_sequence(i):
                invalid_sum += i
            
invalid_sum
```

```python
%%time

input_file = "2.txt"

invalid_sum = 0

def has_repeated_sequence(number):
    num_str = str(number)
    
    for sequence_size in range(1, len(num_str) // 2 + 1):
        sequence = num_str[:sequence_size]

        if sequence * (len(num_str) // sequence_size) == num_str:
            return True

    return False
        
with open(input_file, "r") as file:
    ranges = file.read().strip().split(",")

    for id_range in ranges:
        start = int(id_range.split("-")[0])
        end = int(id_range.split("-")[1])

        for i in range(start, end + 1):
            if has_repeated_sequence(i):
                invalid_sum += i
            
invalid_sum
```

### #5

```python
%%time

input_file = "3.txt"

sum_largest_numbers = 0

def find_largest_two_digits_number(num_str):
    first_digit = -1
    second_digit = -1

    for i in range(len(num_str) - 1):
        if first_digit < int(num_str[i]):
            first_digit = int(num_str[i])
            second_digit = int(num_str[i + 1])
        
        if second_digit < int(num_str[i + 1]):
            second_digit = int(num_str[i + 1])

    return first_digit * 10 + second_digit

with open(input_file, "r") as file:
    for line in file:
        sum_largest_numbers += find_largest_two_digits_number(line.strip())

sum_largest_numbers
```

### #6

```python
%%time

input_file = "3.txt"

sum_largest_numbers = 0

def find_largest_twelve_digits_number(num_str):
    result = 0
    digit = -1
    digit_index = 0

    for i in range(12):
        for j in range(digit_index, len(num_str) - 12 + i + 1):
            if digit < int(num_str[j]):
                digit = int(num_str[j])
                digit_index = j + 1
        
        result += digit * (10 ** (11 - i))
        digit = -1

    return result

with open(input_file, "r") as file:
    for line in file:
        sum_largest_numbers += find_largest_twelve_digits_number(line.strip())

sum_largest_numbers
```
